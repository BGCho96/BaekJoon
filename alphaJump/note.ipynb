# 다이아몬드 수  
n의 복잡도로 다이아몬드 수 방법으로 검증 가능한지 한번 보자.  
n번째 다이아몬드 일 때, 2n+1의 여분이 가능하다면 진행하는 식으로?  
2^31로 경우의 수가 꽤 큰 편이다. 빠르게 구현하고 일반식 등의 방법으로 시간 단축 가능한지 알아보자  
시작도 하기 전에 예외 경우를 발견했다. 다이아몬드 수 가 아니라, 진동감쇄(?) 방법의 접근이 가능하다.  
조건은 모르겠지만, 있는대로 가속을 해서 빠르게 도착한 후, 좌우 반복 감속을 통해서 브레이크가 가능하다. 이런 경우의 수 가 뭐가 있을까....  
일단은 다이아몬드 수 구현부터 하자. 알고리즘 골자부터!  

그냥 좀 싱겁게 끝났다. 다이아몬드 수를 배열하다가 제곱수인 것에 착안, 그냥 주어진 거리보다 작은 제일 큰 제곱수를 뺴고, 나머지를 제곱 레벨보다 낮은 수부터 순차적으로 제거 해나갔다.  
그림으로 설명하자면, 가능한 큰 산을 세운 다음, 긴 기둥부터 가운에서 넣어가면서 목표 숫자를 완성했다.  
조금만 생각해보면, 음수 적분 즉 뒤로 항해를 해서 이득을 보는 구간이 거의 없다. 문제에서 막았기 때문에....  
조금은 싱겁게 풀었다.
