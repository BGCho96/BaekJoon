#알고리즘  
일단은 배낭 알고리즘을 사용해야 하지만, 근본적으로 다른 점은 정확하게 카페인 양을 맞춰야 한다는 점이다.  
배낭의 경우 가방의 용량이 남아도 되지만, 정량의 경우는 다름으로 초반에는 닥치는대로 마신 후, 추후 용량 조정 등의 다른 접근이 필요해 보인다.  
카페인의 양을 딱 맞춘다 -> 최소한을 마신다 라는 추가적인 조건이 있는 상황.  
현재 떠오르는 논리로는 차영이의 커피 용량을 가방의 용량, 남은 가방의 용량을 최소화시키는 점화식을 고려해 볼 수 있다.  
식도 달라졌고 하니, 점화식을 세워보는 것이 좋을 것 같다. 설계는 실행을 압도한다나.  
원래 식은 :  
  
# dp[n][k+1] = max(dp[n][k],dp[n-M[k+1]][k]+V[K+1])  
이 때 M은 용량, V는 가치가 된다. 가방에 담긴 가치의 총량을 dp 값으로, 새로운 대상을 가방에 넣는지 확인하는 것이다.  
이 식을 차영이의 '카페인 배부름 정도', kf의 잔여 용량 식으로 바꾸는 것으로 시작해 보자.  
최대화 되어야 할 '가방 안의 가치의 합' 은 최소화 되어야 할 '마셔야 하는 커피의 잔 수' 로 치환된다. 
# kf[n][k+1] = min(kf[n][k], kf[n-K[k+1]][k]+1)  
논리적 차이로는 이거 한잔 더 마셔서 잔수의 이득을 볼 수 있는가이다. 그럼 조건이 좀 더 추가되는데,  
1. 이 한잔을 마실수는 있는가?(n-K[k+1]>0)  
다른 조건을 붙여보려는데 조건의 확신이 스지 않는다. 결국은 잔여도를 0을 만들어야 하는데, 작은 방향으로 최소 잔 수 와 최소 잔여도 이 두개를 동시에 어떻게 가져가는가, 우선순위로 해결하면 되나, 등....  
리턴값을 배열로 줘버릴까 하는 생각도 든다. 일단은 0~N까지의 배부름 용량, k잔까지의 커피를 고려하는 반복문으로 위 식에 대해 틀을 잡은 뒤 생각해보자.  


2가지 문제점이 생겼다. 사실 한가지 문제인데, 점화식에서 n-1번째와 n번째를 구분하지 못한다.  
그래서 :  
마셨던 커피를 또 마신다. 
k+1번째임으로 2잔을 마셔야 용량을 꽉 차운건데, k번째 경우에서 난 1잔만 마셔도 카페인 달성했어! 와 같은 경우를 처리 못하고 1잔만 마시면 된다고 처리한다.  
그래서  드는 느낌은, 몇잔을 마시면 되느냐 보다는 마신 커피에 대한 인덱스 형식으로 저장을 해야할거 같다.  
아니면 그냥 k번째 물건이 아니라 n번째 용량을 늘리면 되는건가...?  
표를 그려보고 원본 배낭 알고리즘을 확인해야 할 거 같다.
