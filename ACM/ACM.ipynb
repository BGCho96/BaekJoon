# 역순으로 제작 후 진행  
일단은 문제를 찢어서 입력을 받고, 유사 트리 형태로 순서도를 맹그러보자 

재귀...로 짜고싶은데 그러려면 말로 좀 이쁘게 써야겠다.  
1. 끝종단 - 그냥 건설시간을 출력한다. 가장 심플한 로직의 tureCost(끝)의 리턴.  
2. 1개의 자식이 존재. 현재 건설시간 + 자식의 트루코스트를 더한 값.  
3. 다수의 자식이 존재. 현재 건설시간 + 자식들의 트루코스트 중에서 최대값을 더한 값.

재귀로직은 성공적으로 작동한거 같다. 시간초과는 봤어도 틀렸습니다는 안떳으니까. 새로운 개념이 중간에 보여서 그걸 배우고, 도입하기로 했다.  
위상 정렬로, 들어오는 노드의 갯수로 접근하여 진로 탐색과 조건 충족을 동시에 해결했다. 내가 걱정했던 순서를 정하기 - 트리의 완성 관계에서 트리의 완성을 없애줬다.
