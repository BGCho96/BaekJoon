구현이 골치아프고 골드 1난이도를 고려하여 완전탐색을 제하고, 동적을 먼저 시도해보자.  
그냥 떠오른 상향식은, 얼핏 생각하면 중간에 1에 도착하거나 불가능한 경우의 수 일 때 경우의 수가 줄어들지만, 반대로 생각하면 하향식과 동일한데 시작 시드가 10배가 되는 경우다.
