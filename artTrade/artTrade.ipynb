당연히 막히는 경우는 상상하기 쉽다. 요컨대 1번이 3번에게 팔아야만 연쇄적으로 판매가 가능한... 그런데 3번이 3원에 사고 2번이 1원에 사고...  
그러니까 가격변동의 최소화에 의한 추적은 한명의 거래에서 이동이 뚝 끊기는 현상이 발생한다.

시간초과도, 오답도 안뜨지만 메모리 초과.....어떻게 해야 가볍게 가져갈 수 있을까 고민해야겠는데....
