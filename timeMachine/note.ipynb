순환을 한다는 암시를 하는걸까, 아니면... 1을 제외한 대각행렬에서 음수의 값을 가지게 된다는 뜻 아닐까...?

+ 자기 자신의 도시로 돌아가는 것은 순간이동으로 규정 가능하다. 그냥 0을 넣어도 보자 .
