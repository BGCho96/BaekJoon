# 스케일 파악  
감당안되는 숫자를 보고도 일단은 브루탈로 풀었다. 당연히 안풀렸다.  
그러나 정답도 브루탈이다. 심지어는 "소수는 노가다 아니면 안된다" 를 증명하는 브루탈의 시초같은 방법인 에라토스테네스의 체의 응용이었다.  
내가 놓친 부분을 정리해보자면 :  
체를 두고 밀가루를 한알한알 옮기는 발상을 한 것이다.  
어떤 범위 안의 제곱수 약수 거르기 - 범위 최대의 제곱근 만큼 반복.  
어떤 범위 안의 제곱수 일일히 거르기 - N제곱 시간 복잡도 k 회 반복.  
둘다 막연해보이지만 한쪽이 상한이 정해져 있는 스마트한 방법이다. 뭐라 말로 표현 못하겠지만 나는 두 삽 중에 어느 삽이 더 무거운지 읽지 못한 것이다.  
조금 정리를 하고 축약해보자 노력한다면, 이 문제는 가장 복잡한 경우의 수(가장 큰 경우) 안에 하위 문제에 대한 모든 풀이를 포함하고 있다. 그 방법을 확장하고자 했어야 했다.
