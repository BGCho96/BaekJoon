# 일반식 세우기  
어떤 입력에도 효율적으로 출력하려면, 아무래도 일반식이 맞는거 같다.  
토네이도를 잘 그려서, 원점 기준으로 X 자를 그리면 4사분면을 제외하고는 모서리들이 이쁘게 잘 걸린다. 그말은 꼭지점의 패턴을 이용하면 일반식 가능하다는게 아닐까?  
일단은 새로운 4분면(상하좌우)에 대해서 일반 조건식을 세워보자.  
![poster](./image.png)  
일단 상 :  
좌우보다 위로 뻗어야 하니,  
abs(x)<-y  
하는 반대로,  
abs(x)<y
좌의 경우,  
abs(y)<-x  
우는,  
abs(y)<x  
단, 우측의 경우 엄밀히 말하면 하에 해당되는 대각선을 포함한다. 그러니,  
if abs(y)<x, and x==y+1의 경우, 하단의 경우의 수로 보내버린다.  
그럼 나는 이 4가지 경우의 수를  놓고, 일반식을 써보는걸 구현해보겠다.  

무식하게 해서 맞긴 했는데,... 소요시간을 생각하면 좀 뻘짓이 있기도 했다.
