# square first  
해당 문제에서 가장 클 수 있는 자릿수는 대각 갯수일거다.  
그러니까 그때까지만 square를 돌려보고, 지정된 숫자 배열을 표 안에서 찾아보자.  
애초에 문제에서 표는 최대 9*9 환경이다. 9자리가 최대인데, 10000개 정도 돌려보는거면,,,,할만한 선이라고 생각된다.  
그럼 일단은 지정된 숫자가 몇개 있는 세보는걸로 가자.  
그럴려면 일단은 2*2 배열과 별개로, 각 숫자 요소(0~9)가 몇개씩 분포되어 있는지도 알면 좋을거 같다. 아님 아에 숫자별로 좌표 정보를 따로 set로 저장해보는 것도 좋겠다.
